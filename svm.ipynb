svc() parameters 
kernel is Kernel coefficient for ‘rbf’, ‘poly’ and ‘sigmoid’
c-Regularization parameter. The strength of the regularization is inversely proportional to C. 

gama is kernal coefficients for rbf,poly
if gamma='scale' (default) is passed then it uses 1 / (n_features * X.var()) as value of gamma
if ‘auto’, uses 1 / n_features
if float, must be non-negative.


```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
df=pd.read_csv('/content/heart.csv')
df
```

```python
def fitting(x,y,C,gamma):
   x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
   iters=['rbf','poly','sigmoid']

   for i in iters:
     model=SVC(kernel=i,probability=True,C=C,gamma=gamma)
     clf=model.fit(x_train,y_train)

     pred_train=model.predict(x_train)
     pred_test=model.predict(x_test)

     #use score to get accuracy of model
     #test_score=model.score(x_test,y_test)
     print("accuracy score: ",accuracy_score(y_test,pred_test))
     matrix=metrics.confusion_matrix(y_test,pred_test) 
     print("confusion_matrix \n",matrix)
     print(classification_report(y_test,pred_test,zero_division=0))
     

      #evaluate train model
     train_score=model.score(x_train,y_train)
     print("accuracy score: ",train_score)

     print(classification_report(y_train,pred_train,zero_division=0))



     


```

```python
x=df.drop('target',axis=1)
y=df['target'].values
fitting(x,y,1,'scale')
```
