```python
def get_mean(numbers):
      sum=0
      i=0
      n=len(numbers)
      for i in numbers:
        sum=sum+i

      mean=sum/n
      return mean
```

```python
def get_mode(numbers):
   max_num=-1

   occurances={}
   for i in numbers:
     if(occurances.get(i)==None):
         occurances[i]=1
     else:
         occurances[i]+=1

     if occurances[i]>max_num:
       max_num=occurances[i]
     
   for i in occurances:
     if occurances[i]==max_num:
           return i
   
   return none
     
```

```python
def get_median(features):
   features.sort()

   n=len(features)
   mid=n//2
   print(features)
   return features[mid]
```

```python
def get_sd(features):
    mean=get_mean(features)
    n=len(features)
    std=0

    for i in features:
      std+=(i-mean)**2

    sd=(std/n)**0.5
    return sd
```

```python
def normalization(features):
    x_max=max(features)
    x_min=min(features)
    
    scaled=[]

    for i in features:
      scaled.append((i-x_min)/(x_max-x_min))

    return scaled

    
```

```python
def std_values(features):
   std_val=[]
   mean=get_mean(features)
   sd=get_sd(features)

   for i in features:
       std_val.append((i-mean)/sd)

   return std_val
```

```python
def variance(features):
    
    sd=get_sd(features)
    return sd**2
```

```python
features=[100000,-2,50,12,700,9000,50,50,12,700,50,12]
print("mean: ",get_mean(features))
print("mode: ",get_mode(features))
print("median: ",get_median(features))
print("sd: ",get_sd(features))
print("normalization: ",normalization(features))
print("std_values: ",std_values(features))
print("var: ",variance(features))
```
